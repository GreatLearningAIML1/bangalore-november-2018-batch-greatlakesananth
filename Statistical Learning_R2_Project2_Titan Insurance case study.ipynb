```python
import pandas as pd
import random
import numpy as np
import statistics
from scipy import stats
from math import sqrt
from scipy.stats import t
from scipy.stats import shapiro
bc_df = pd.read_csv("Titan_assignment.csv")
bc_df.Old_Scheme_in_Thousand = bc_df.Old_Scheme_in_Thousand*1000
bc_df.New_Scheme_in_Thousand = bc_df.New_Scheme_in_Thousand*1000
```

```python
μ1 = bc_df.Old_Scheme_in_Thousand.mean()
print('Mean of Old Scheme is',μ1)
```

```python
μ2 = bc_df.New_Scheme_in_Thousand.mean()
print('Mean of New Scheme is',μ2)
```

```python
# check for normality in the data.
titandiff = bc_df.New_Scheme_in_Thousand - bc_df.Old_Scheme_in_Thousand
titandiff.plot(kind='hist', title= 'Titan Histogram')
stat, p = shapiro(titandiff)
print('Since p value', p, 'is greater than alpha 0.05 we assume normal distribution')
```

```python
#Null Hypothesis : There is no significant difference in sales with the new scheme
#Alternate Hypothesis : There is a significant difference in sales with the new scheme
#μ1 = Average sums assured by salesperson BEFORE changeover. μ2 = Average sums assured by salesperson AFTER changeover.
#H0: μ1 = μ2  ; μ2 – μ1  = 0
#HA: μ1 < μ2   ; μ2 – μ1  > 0 ; true difference of means is greater than zero.
from scipy.stats import ttest_rel
N = len(bc_df) #this is the sample size 
df = N-1
tstat,p = ttest_rel(bc_df.New_Scheme_in_Thousand, bc_df.Old_Scheme_in_Thousand)
print('TStat Value is ',tstat)
pvalue = 1-(stats.t.cdf(tstat,df))
print('P Value is ',pvalue)
print ('Since  0.0652(P) > 0.05(alpha), we accept (fail to reject) NULL hypothesis. The new scheme has NOT significantly raised outputs.')
```

```python
# What is the p-value of the hypothesis test if we test for a difference of $5000
#Let  μ2 = Average sums assured by salesperson AFTER changeover.

#μ1 = Average sums assured by salesperson BEFORE changeover.

#μd = μ2 – μ1   H0: μd ≤ 5000  HA: μd > 5000
alpha = 0.05
print('alpha which is the probability of a type 1 error is 0.05')
```

```python
#H0(null hypothesis): μd ≤ 5000  HA(Alternate Hypothesis): μd > 5000
xbar_1 = μ1
mu = xbar_1 + 5000
sd = statistics.stdev(bc_df.Old_Scheme_in_Thousand)
n = 30
se = sd/sqrt(n)
z = (mu-xbar_1)/se
#This is a right tail test.
p = 1- stats.norm.cdf(z)
print ('P value is',p)
print ('Since',p,'(p)is > than 0.05(alpha)we accept (fail to reject) NULL hypothesis')
```

```python
# define critical value T.
p = 1- alpha
value = t.ppf(p, df)
print('Critical Value of T is',value)
```

```python
μ0 = 0
μd = 5000
titandiff = bc_df.New_Scheme_in_Thousand - bc_df.Old_Scheme_in_Thousand
SE = statistics.stdev(titandiff)/sqrt(N)
xbar = μ0 + value*SE
t_stat = (xbar-μd)/SE
p = (stats.t.cdf(t_stat,df=df))
print('Power of Test is',p)
beta = 1-p
print('Beta which is a type 2 error is',beta)
```
